```python
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
import numpy as np
import folium
from IPython.display import display
matplotlib.style.use('ggplot')
```

```python
%matplotlib inline
years = ['2009', '2010', '2011', '2012', '2013', '2014', '2015']
n2 = [0]*7
i = 0
for year in years:
    df = pd.read_csv('data2/Accidents_SelectedRoad' + year + '.csv', low_memory=False)
    n2[i] = len(df)
    i += 1

x = range(7)
plt.figure(figsize = (15,6))
plt.subplot(1,2,1)
plt.bar(x, n2)
index = [0.5,1.5,2.5,3.5,4.5,5.5, 6.5]
years = ['2009', '2010', '2011', '2012', '2013', '2014', '2015']
plt.xticks([0.5,1.5,2.5,3.5,4.5,5.5,6.5], years)
#plt.title('Total number of accidents on Motorway and A(M) in each year')
for j, v in zip(index, n2):
        plt.text(j, v+100, v, ha='center', va='bottom',fontsize = 10)
plt.savefig('image/accident_num_highways_allyear.eps',format="eps",transparent=True, bbox_inches='tight', pad_inches=0.1)
plt.show()
```

```python
%matplotlib inline
years = ['2009', '2010', '2011', '2012', '2013', '2014', '2015']
nfig = 1
plt.figure(figsize = (10,10))
for year in years:
    df = pd.read_csv('data2/Accidents_' + year + '.csv',low_memory=False)
    date = df['Date'].values
    n = len(date)
    month = pd.Series([0]*n)
    i = 0
    for d in date:
        m = d.split('/')[1]
        month[i] = m
        i+=1
    months = month.value_counts()
    months = months.reindex(range(1,13))
    plt.subplot(3,3,nfig)
    plt.bar(months.index, months.values,align = 'center')
    plt.title(year)
    nfig += 1
plt.savefig('image/accident_num_highways_month_allyear.eps',format="eps",transparent=True, bbox_inches='tight', pad_inches=0.1)
plt.show()
```

```python
df = pd.read_csv('data2/Accidents_SelectedRoad2014.csv',low_memory=False)
times = df['Time'].values
n = len(times)
hour = pd.Series([0]*n)
i = 0
for t in times:
    h = t.split(':')[0]
    hour[i] = h
    i+=1
hours = hour.value_counts()
hours = hours.reindex(range(24))
plt.figure(figsize = (5,4))
plt.bar(hours.index, hours.values,align = 'center')
#plt.title('the number of accident in each hour in a day in 2014')
plt.savefig('image/accident_num_highways_hour_2014.eps',format="eps",transparent=True, bbox_inches='tight', pad_inches=0.1)
```
